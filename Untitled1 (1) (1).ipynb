```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
well_data=pd.read_csv("heatmap.csv", encoding = 'utf-8')
well_data=well_data[['Comp Score','NYSE','NASDAQ','SSE','HSE','LSE']]
```

```python
well_data
```

```python
def convert(i):
    return float(str(i.replace(",","")))

for i in well_data.columns[1:]:
    well_data[i] = well_data[i].apply(convert)
```

```python
cor=well_data.corr()
```

```python
cor
```

```python
sns.heatmap(cor,cmap='RdBu',vmin=-1,vmax=1,annot=True)
```
